```python
n, p = 1, 0.5 #n=number of trails,p=probability of each trail
s = np.random.binomial(n,p,1000)
plt.hist(s)
```

```python
from scipy.stats import bernoulli
import seaborn as sns
bernoulli_data = bernoulli.rvs(size=1000,p=0.6)
ax = sns.distplot(bernoulli_data,
                kde=True,
                color='crimson',
                hist_kws = {'linewidth':25,
                           'alpha':1})
ax.set(xlabel='Bernoulli',ylabel='Frequency')
```

# Geometric Distribution

```python
from __future__ import print_function, division
from scipy.stats import geom
```

```python
prob = 0.2
n = 12
p = np.zeros(n)
q = np.zeros(n)
for k in range(1,n+1):
    p[k-1] = (1-prob)**(k-1)*prob
    q[k-1] = geom.pmf(k,prob)
plt.bar(range(1,13),p,color='green')
plt.xlim(1,15)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title("First Success at n")
plt.show
```

# Normal Distribution

```python
mu,sigma = 0.5, 0.1
s = np.random.normal(mu,sigma,1000)
count,bins,ignored = plt.hist(s,20)
plt.plot(bins,1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2/(2*sigma**2)),linewidth=3,color='y')
plt.show()
```

# Central Limit Theorem and Sampling Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# population of size 10000
population = np.random.binomial(10,0.4,10000)
```

```python
sample_means = []
for x in range(1,21):
    sample = np.random.choice(population,size=int(0.3*len(population)),replace=False)
    sample_means.append(np.mean(sample))
```

```python
# Calculating mean of sample means and population mean
mean_of_sample_means = np.mean(sample_means)
population_mean = np.mean(population)

print(f'Mean of Sample Means {mean_of_sample_means}')
print(f'Mean of Population {population_mean}')
```

```python
ax = sns.distplot(sample_means,kde=True,color='orange',hist_kws={'linewidth':15,'alpha':1})
```

# Cumulative Distribution Function

```python
import scipy.stats as ss
x = np.linspace(-5,5,5000)
mu = 0
sigma = 1

y_pdf = ss.norm.pdf(x,mu,sigma)
y_cdf = ss.norm.cdf(x,mu,sigma)

plt.plot(x,y_pdf,label='pdf')
plt.plot(x,y_cdf,label='cdf')
```
